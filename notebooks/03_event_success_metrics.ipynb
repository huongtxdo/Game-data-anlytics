Scenario: A time-limited in-game event where players can earn unique rewards

Context: Players complete levels to win exclusive items, bonuses, or coins. 

Let define some metrics and KPIs to determine the success of the event!

Participation rate: Percentage of players taking part in the event:
$$
\frac{\text{total players participating in the event at least once}}{\text{total players logging in game}}
$$
=> This metrics together with the average number of unique login players before and during the event tell us how efficient the event is in attracting players. understanding an event's popularity helps us know where to focus our resources on, and possibly develop monetizing strategies upon it.

Level completion:
$$
\frac{\text{sum of all levels reached}}{\text{total players logging in game}}
$$
=> This metrics determines how much players are engaged in the event. Less level completion poses the problem of the event being too hard or boring, knowing which helps bring timely adjustment to the event. 

Average attempts made per level:
$$
\frac{\text{total numbers of attemps made per level}}{\text{total players logging in game}}
$$
=> This serves as a base for figuring out which levels interest players most, or whether the levels' difficulty needs toned down.
